```python
fruits = ['orange'] * 5 + ['apple'] * 4

experiments = 1000000

succes = 0

for _ in tqdm(range(experiments)):
    selected_fruits = random.sample(fruits, 3)

    if selected_fruits.count('orange') == 3:
        succes += 1

probability = succes / experiments

print("Вероятность выбрать все три апельсина:", probability)

if round(probability,3) == 0.012:
   print("Правильно")
else:
   print("Не подтверждено")
```

```python
import random
from tqdm import tqdm

detail= ['norm'] * 7 + ['invalid'] * 3

experiments = 1000000

succes = 0

for _ in tqdm(range(experiments)):
    selected_detais = random.sample(detail, 2)

    if selected_detais[0] == 'norm' and selected_detais[-1] == 'invalid':
        succes += 1
probability = succes / experiments

print("Вероятность выбрать только 2 детали:", probability)

if round(probability,3) == 0.023:
   print("Правильно")
else:
   print("Не подтверждено")
```
